```python
# reqeusts, bs4 import
import requests
# BeautifulSoup 클래스 import
import bs4
# beautifulsoup 클래스 import
from bs4 import BeautifulSoup
```

```python
#01_news_scraping.ipynb

# 경제 뉴스 
req_param = {
    'sid': 'economy'
}


url = 'https://news.daum.net/{sid}'.format(**req_param)
print(url)

# 요청 헤더 설정 : 브라우저 정보
req_header = {    
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36"
}
# requests 의 get() 함수 호출하기 
res=requests.get(url, headers=req_header)
res.encoding = 'utf-8'
print(type(res))
print(res.status_code)

if(res.ok):
    html = res.text
    # BeautifulSoup 객체 생성  
    soup = BeautifulSoup(html, "html.parser")


li_tags = soup.select('ul.list_newsheadline2 li')
print(type(li_tags),len(li_tags))

idx = 1
for li in li_tags:
    a_tag = li.find('a')
    if not a_tag:
        continue
    link = a_tag.get('href')
    strong_tag = li.select_one('div.cont_thumb strong.tit_txt')
    if strong_tag:
        title = strong_tag.get_text(strip=True)
        if title and link:
            print(link)
            print(title)
            idx += 1
```

```python
import requests
from bs4 import BeautifulSoup

section_dict = {'기후/환경' : 'climate', '사회' : 'society', '경제': 'economy', '정치' : 'politics',
                '국제' : 'world', '문화' : 'culture', '생활' : 'life', 'IT/과학' : 'tech', '인물' : 'people'}

def print_news(section):  #print_new(103,'생활/문화') 
    sid= section_dict[section]
    url = f'https://news.daum.net/{sid}'
    print(f'=====> {url} {section} 뉴스 <=====')

    req_header = {    
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36"
    }
    # requests 의 get() 함수 호출하기 
    res=requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if(res.ok):
        html = res.text
        # BeautifulSoup 객체 생성  
        soup = BeautifulSoup(html, "html.parser")


    li_tags = soup.select('ul.list_newsheadline2 li')

    idx = 1
    for li in li_tags:
        a_tag = li.find('a')
        if not a_tag:
            continue
        link = a_tag.get('href')
        strong_tag = li.select_one('div.cont_thumb strong.tit_txt')
        if strong_tag:
            title = strong_tag.get_text(strip=True)
            if title and link:
                print(link)
                print(title)
                idx += 1
```

```python
print_news('경제')
print_news('사회')
```
